```python
import numpy as np

import math

import pdb

import torch

from collections import namedtuple
```

```python
class linkedList():
    
    def __init__(self, value):
        
        self.value = value
        
        self.next = None
```

```python
# creating linked list

data = np.arange(10)
        
a = linkedList(data[1])

b = linkedList(data[2])

c = linkedList(data[7])

d = linkedList(data[4])

e = linkedList(data[5])

a.next = b

b.next = c

c.next = d

d.next = e
```

```python
class Operation:
    
    def __init__(self):
        
        pass
    
    def insert(self, head, LinkedList, index, value):
        
        current_node = head
        
        for iteration in range(index - 2):
            
            current_node = current_node.next
            
        next_ = current_node.next
        
        new_node = LinkedList(value)
            
        current_node.next = new_node
        
        new_node.next = next_
            
    def delete(self, head, index):
        
        current_node = head
        
        for iteration in range(index - 2):
            
            current_node = current_node.next
            
        current_node.next = current_node.next.next
    
    def reverse(self, head):
        
        prev = None
        
        current = head
        
        while current != None:
            
            next_ = current.next
            
            current.next = prev
            
            prev = current
            
            current = next_
    
    def findValue(self, head, index):
        
        current_node = head
        
        count = 0
        
        while head != None:
            
            if count == index - 1:
                
                return current_node.value
            
            current_node = current_node.next
            
            count += 1
    
    def findIndex(self, head, value):
        
        current_node = head
        
        count = 0
        
        while current_node != None:
            
            if current_node.value == value:
                
                return count + 1
            
            count += 1
            
            current_node = current_node.next
                
        return "no value"
```

```python
linkedlist = Operation()

### insert element

linkedlist.insert(head = a, LinkedList = linkedList, index = 3, value = 10)

## delete element

linkedlist.delete(head = a, index = 3)

### find value by index

linkedlist.findValue(head = a, index = 3)

### find index by value

linkedlist.findIndex(head = a, value = 7)

### reverse

linkedlist.reverse(head = a)
```
